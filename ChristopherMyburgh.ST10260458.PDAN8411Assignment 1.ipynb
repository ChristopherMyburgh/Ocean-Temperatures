# Linear Regression


Linear regression is a statistical method used to analyze the relationship between a dependent variable (often denoted as y) and one or more independent variables (often denoted as x). The goal of linear regression is to find the line that best fits the data, which is often referred to as the "line of best fit."

The line of best fit is the one that minimizes the distance between the predicted values of the dependent variable and the actual values of the dependent variable. In other words, linear regression seeks to find the line that best describes the relationship between the dependent variable and the independent variable, such that the predicted values of the dependent variable are as close as possible to the actual values. It's important to note that the line of best fit is only an approximation of the true relationship between the variables. There may be other factors that influence the relationship that are not captured by the line of best fit.

According to "The Elements of Statistical Learning" by Friedman, Tibshirani, and Hastie, the dependent variable in linear regression is usually continuous, meaning that it can take on any value within a range (e.g., age, height, weight, income). The independent variables can be either continuous or categorical, meaning that they can take on a limited number of values (e.g., gender, education level, marital status)(Friedman et al., 2009)..

There are two types of linear regression: simple linear regression and multiple linear regression. Simple linear regression involves only one independent variable, while multiple linear regression involves two or more independent variables.

For Simple Linear Regression; The formula can be expressed as:

y = β0 + β1x + ε

where y is the dependent variable, x is the independent variable, β0 is the y-intercept or bias term, β1 is the slope of the line or weight term, and ε is the error term representing the residual or unexplained variance.

For Multiple Linear Regression; The formula can be expressed as:

y = β0 + β1x1 + β2x2 + ... + βpxp + ε

where y is the dependent variable, x1, x2, ..., xp are p independent variables, β0 is the y-intercept or bias term, β1, β2, ..., βp are the slopes of the line or weight terms for each independent variable, and ε is the error term representing the residual or unexplained variance.

The goal of multiple linear regression is to estimate the values of the coefficients β0, β1, β2, ..., βp that minimize the sum of squared errors between the predicted values and the actual values of the dependent variable.

In order to perform linear regression, we need a dataset that includes values for the dependent variable and the independent variable(s). We use this data to calculate the coefficients of the line of best fit, which represent the slope and intercept of the line.

Once we have calculated the coefficients of the line of best fit, we can use the equation for the line to make predictions about the dependent variable based on values of the independent variable(s). These predictions are called "predicted values" or "fitted values," and they are represented by the line on a graph of the data.

The goals of linear regression in data analysis can vary depending on the context, but there are some common goals that are often pursued:

__Prediction__: Linear regression can be used to predict the value of the dependent variable based on the value of the independent variables.

__Understanding relationships__: Linear regression can help us understand how the dependent variable is related to the independent variables.

__Testing hypotheses__: Linear regression can be used to test hypotheses about the relationship between variables. For example, we might want to test the hypothesis that there is a significant relationship between two variables, or that the relationship between two variables is different for different groups of people.

__Control for confounding variables__: Linear regression can be used to control for confounding variables, which are variables that can influence the relationship between the independent variables and the dependent variable.

It's worth noting that linear regression assumes a linear relationship between the dependent variable and the independent variable. If the relationship is not linear, linear regression may not be the best method to use.

## Missing Data Imputation
My plan for imputing the missing data involves comparing the mode and mean values of each column before and after removing the null values. If a column has a mode of NaN when comparing the mode with null values and a different mode when comparing without, I will use the mean of the modes to impute the missing data. However, it's important to note that the mode and mean values only provide information about the distribution of individual variables, and do not capture relationships between variables.

```python
# Create a for loop that will display each columns Mode and Mean value with missing data vs without missing data

for i in range(df.shape[1]):
    column = df.iloc[:, i]
    mode_value_with_nan = stats.mode(column, keepdims=False)
    mode_value_without_nan = stats.mode(column, nan_policy='omit', keepdims=False)
    mean_value_with_nan = np.mean(column)
    mean_value_without_nan = np.nanmean(column)
    print(f"Column {i+1}:\n Mode without NaN = {mode_value_without_nan},\n Mode with NaN = {mode_value_with_nan},\n Mean without NaN = {mean_value_without_nan},\n Mean with NaN = {mean_value_with_nan}")
```

Based on the above, we can see that some columns have a mode of missing data. This tells us that the most common occuring value was a NULL value. What is interesting to note, is that while the mode changes, the mean of the data remains relatively the same, with and without missing values. This tells me that we may apply mean imputation, where we fill null values with the mean of the feature.

```python
# Create a for loop that imputes the missing data with the mean of each column

for i in df.columns:
    if df[i].isnull().sum() > 0:
        df[i].fillna(df[i].mean(), inplace=True)
        print('filled',i)
```

## Normalization
One of the key assumptions that Regression analysis makes, is that the data is normally distributed(Friedman et al., 2009).. In order for us to perform any kind of regression analysis, we must first implement normalization on our data.<br>I am going to be using the StandardScaler method from sklearn due to the following reasons:

<br>- Linear regression assumes that the input features are normally distributed: Linear regression models make an assumption that the input features are normally distributed. StandardScaler scales the features so that they have a mean of 0 and a standard deviation of 1, which helps to ensure that the features are closer to a normal distribution. This can improve the performance of linear regression models.
<br>- StandardScaler reduces the impact of outliers: Outliers in the input data can have a large effect on the regression coefficients and make the model less accurate. StandardScaler reduces the impact of outliers by scaling the features based on their standard deviation. This means that outliers will be scaled down and have less influence on the regression coefficients.
<br>- StandardScaler makes it easier to compare the importance of different features: When using StandardScaler, the feature values are on the same scale, which makes it easier to compare the importance of different features in the regression model. This is because the coefficients can be interpreted as the change in the output variable that is associated with a one standard deviation change in the corresponding input feature.

# Regression Analysis and Model Performance

## Simple Multiple Linear Regression

### Choosing Variables

When performing Linear Regression, we know that we need to first normalize our data in order to reduce the magnitude each variable might have on the target feature, to ensure that variables with different scales are treated equally. If the independent variables have different scales, they may have a different impact on the dependent variable, even if they are equally important in explaining it. Normalizing the variables puts them on a common scale, which helps to avoid this problem and ensures that they are treated equally in the regression model.<br>
<br>Therefore, my X variable (Independent Variable) will consist of the features that have been normalized.<br>
<br> My y variable (Target Variable) is the variable that I am trying to predict. In this case, it will be the Temperature of the water. When performing Linear Regression, we do not need to worry about normalizing our target variable. Considering this, our target variable will be the feature as it is in its true state.

```python
# Selecting all features from column 2 - 11 on the normalized dataset
X = df_normal.iloc[:,1:11]

# Selecting the first column to be the target variable
y = df['T_degC']
```

```python
# Looking at the shape of my variables
print(X.shape)
print(y.shape)
```

```python
# split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle = True, test_size=0.2, random_state=42)
    
# fit a linear model to the training data
model = LinearRegression().fit(X_train, y_train)
    
# make predictions on the test data
y_pred = model.predict(X_test)
    
# plot the predicted values against the actual values
plt.scatter(y_test, y_pred, s=1, c='red')
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Multiple Linear Regression")
plt.show()
```

## Model Evaluation

```python
# R-squared score
print('R2 Score of our model on the Test Datatset: ',r2_score(y_test, y_pred))

# Mean absolute error
print('\nMean Absolute Error Score of our model on the Test Datatset: ',mean_absolute_error(y_test, y_pred))

# Mean Squared error
print('\nMean Squared Error Score of our model on the Test Datatset: ',mean_squared_error(y_test, y_pred))
```

R2 Score: This metric represents the proportion of variance in the target variable that is explained by the independent variables. An R2 score of 0.9716 indicates that the model is able to explain 97.16% of the variance in the target variable using the independent variables. This is a high R2 score, which suggests that the model is a good fit for the data.

Mean Absolute Error (MAE): This metric represents the average absolute difference between the predicted and actual values of the target variable. A MAE score of 0.3480 indicates that on average, the model's predictions are off by about 0.3480 units from the actual values of the target variable. This is a relatively small error, which suggests that the model is making accurate predictions.

Mean Squared Error (MSE): This metric represents the average squared difference between the predicted and actual values of the target variable. An MSE score of 0.5088 indicates that on average, the model's predictions are off by about 0.5088 squared units from the actual values of the target variable. This is also a relatively small error, which suggests that the model is making accurate predictions.

Overall, one may conclude that the model is performing well on the test dataset, based on these evaluation metrics. However, it's important to keep in mind that these metrics are only one way to evaluate the performance of a model.

```python
# calculate the percentage of predictions that fall within 10% of the actual value
error_range = 0.05
error = np.abs((y_test - y_pred) / y_test)
within_range = error < error_range
percent_within_range = np.sum(within_range) / len(y_test) * 100
percent_not_within_range = 100 - percent_within_range

print("Percentage of predictions within", error_range*100, "% of actual value:", percent_within_range)
print("\nPercentage of predictions not within", error_range*100, "% of actual value:", percent_not_within_range)
```

```python
# get the coefficients and feature names of the model
coefficients = model.coef_
feature_names = X_train.columns.tolist()

# sort the coefficients and feature names in descending order
sorted_indices = abs(coefficients).argsort()[::-1]
sorted_coefficients = coefficients[sorted_indices]
sorted_feature_names = [feature_names[i] for i in sorted_indices]

# plot the sorted coefficient magnitudes
plt.figure(figsize=(8,6))
plt.bar(range(len(coefficients)), sorted_coefficients)
plt.xticks(range(len(coefficients)), sorted_feature_names, rotation=90)
plt.xlabel("Features")
plt.ylabel("Magnitude")
plt.title("Sorted Coefficient Magnitudes")
plt.show()
```

```python
max_len = max(len(name) for name in sorted_feature_names) # Get maximum length of feature names
for name, coef in zip(sorted_feature_names, sorted_coefficients):
    padding = ' ' * (max_len - len(name) + 5) # Add padding to align columns
    print(name + padding + str(coef))
```

High multicollinearity can lead to unstable coefficients and inflated standard errors, which can reduce the accuracy of the model and make it difficult to interpret the importance of each independent variable. In addition, the coefficients may not accurately represent the true relationship between the independent variables and the dependent variable, as they may be reflecting the effects of multicollinearity rather than true effects.<br>
If multicollinearity is present, I can either consider removing one of the variables or applying techniques such as ridge regression to address it.

## Ridge Regression

### Performing GridSearchCV To Select the best Hyperparameters For Ridge Regression

GridSearchCV is a library from sklearn that allows us to input the different hyperparameters our Ridge model needs to asses. It then iterates through each model and selects the best model based on the penalty (alpha) and the solver applied. 

```python
# split the data into training and testing sets
X_train_R, X_test_R, y_train_R, y_test_R = train_test_split(X, y, shuffle=True, test_size=0.2, random_state=42)

# Define parameters to be tested
parameters = {'alpha': [10, 1, 0.1, 0.01, 0.001, 0.0001] , "fit_intercept": [True, False], 
                "solver": ['svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga']}

# Define the model
model = Ridge()

# Define the grid search
grid_ridge_reg = GridSearchCV(model, parameters, scoring= 'neg_mean_squared_error', cv= 5)

# Fit the grid search with the training data
grid_ridge_reg.fit(X_train_R, y_train_R)

# Best combination of parameters
best_model = grid_ridge_reg.best_estimator_
print(best_model)

# Prediction of the best model
best_pred = best_model.predict(X_test_R)

# PLotting the Predicted vs Actual Values using the best Hyperparameters
plt.scatter(y_test_R, best_pred, s=1, c='red')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title("Ridge Ression using the Best Hyperparamters")
plt.show()
```

## Model Evaluation

```python
# R-squared score
print('R2 Score of our model on the Test Datatset: ',r2_score(y_test_R, best_pred))

# Mean absolute error
print('\nMean Absolute Error Score of our model on the Test Datatset: ',mean_absolute_error(y_test_R, best_pred))

# Mean Squared error
print('\nMean Squared Error Score of our model on the Test Datatset: ',mean_squared_error(y_test_R, best_pred))
```

Looking at the above metrics, The Model performs almost exactly the same using Ridge Regression as it did when applying normal Linear Regression.

### What Now?

The last model I will be performing is a linear regression model that utilizes the net elasticity penalty. The model is called ElasticNet and it is a regularization technique that combines both L1 and L2 penalties to achieve a balance between sparsity and smoothness in the model's coefficients. The two main hyperparameters of the ElasticNet model are alpha and l1_ratio.

alpha controls the strength of the regularization penalty, where higher values of alpha will result in more regularization and sparser coefficient estimates. l1_ratio controls the trade-off between the L1 and L2 penalties, with values of 1.0 resulting in Lasso regression (L1 penalty only), values of 0.0 resulting in Ridge regression (L2 penalty only), and values in between resulting in a combination of the two penalties.

## Net Elasticity Regression

### Performing GridSearchCV To Select the best Hyperparameters For ElasticNet Regression

```python
# I will be importing the warnings library as when using ElasticNet, One may expect to receive deprecating warnings if the models
# accuracy is below the desired threshold
import warnings
warnings.filterwarnings('ignore')


# split the data into training and testing sets
X_train_EN, X_test_EN, y_train_EN, y_test_EN = train_test_split(X, y, shuffle = True, test_size=0.2, random_state=42)

# Define hyperparameters to search over
params = {'alpha': [0.001, 0.01, 0.1, 1, 10], 'l1_ratio': [0.1, 0.3, 0.5, 0.7, 0.9], 'fit_intercept': [True, False]}

# Define the model
E_net_model = ElasticNet()

# Perform grid search with cross-validation
grid_EN = GridSearchCV(E_net_model, params, cv=5)
grid_EN.fit(X_train_EN, y_train_EN)

# Best combination of parameters
best_EN_model = grid_EN.best_estimator_
print(best_EN_model)

# Prediction of the best model
best_EN_pred = best_EN_model.predict(X_test_EN)

# PLotting the Predicted vs Actual Values using the best Hyperparameters
plt.scatter(y_test_EN, best_EN_pred, s=1, c='red')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title("ElasticNet Model using the Best Hyperparamters")
plt.show()
```

## Model Evaluation

```python
# R-squared score
print('R2 Score of our model on the Test Datatset: ',r2_score(y_test_EN, best_EN_pred))

# Mean absolute error
print('\nMean Absolute Error Score of our model on the Test Datatset: ',mean_absolute_error(y_test_EN, best_EN_pred))

# Mean Squared error
print('\nMean Squared Error Score of our model on the Test Datatset: ',mean_squared_error(y_test_EN, best_EN_pred))
```

The above is Very Interesting.

Elastic Net is a regularization method that combines L1 and L2 penalties, whereas Ridge regression only applies L2 penalty. L1 penalty has a tendency to drive coefficients to exactly zero, resulting in a sparse model. In contrast, L2 penalty shrinks the coefficients towards zero but not to exactly zero, so all the features remain in the model, although their coefficients may be very small.

Because of this difference, Elastic Net may produce different magnitudes of coefficients than Ridge regression, especially if there is a high degree of multicollinearity in the data. Elastic Net can also help to select the most important features by driving some of the coefficients to exactly zero.
